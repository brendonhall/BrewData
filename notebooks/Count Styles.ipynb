```python
%matplotlib inline
import matplotlib.pylab as plt
import pandas as pd
import xmltodict

pd.options.display.max_rows = 12
```

```python
xml_file = '/Users/bhall/Projects/BrewData/data/Beer2k.xml'
with open(xml_file) as fd:
    doc = xmltodict.parse(fd.read())
```

```python
counter = 0
styles = {}
for recipe in doc['RECIPES']['RECIPE']:
    counter += 1
print('There are %d beers in the database'%counter)
```

```python
counter = 0
styles = {}
for recipe in doc['RECIPES']['RECIPE']:
    style = recipe['STYLE']['NAME']
    try:
        styles[style] += 1
    except KeyError:
        styles[style] = 1

styles_df = pd.DataFrame(data=styles.values(), index=styles.keys(), columns=['count'])
styles_df.sort_values(inplace=True, by=['count'], ascending=False)
```

```python
styles_df.ix[:30].plot(kind='bar', figsize=(18, 5), title='Style Popularity by Count')
```
